# Bi-variate analysis

Bivariate Analysis of salary and 10percentage

```python
sns.scatterplot(x='10percentage',y='Salary',data=df)
plt.title('salary vs salary')
```

Bivariate analysis of  salary and 10percentage

```python
plt.hexbin(df['10percentage'],df['Salary'],gridsize=30,cmap='Blues')
plt.colorbar(label='count in bin')
plt.xlabel('10percentage')
plt.ylabel('Salary')
plt.title('Hexbin Plot: 10percentage vs Salary')
plt.show()
```

```python
sns.pairplot(df[['10percentage','Salary']])
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='CollegeState', y='Salary', hue='Gender', estimator=sum)  # or estimator=mean
plt.xticks(rotation=90)  # Rotate x labels for better visibility
plt.title('Barplot of Salary by Gender and Collegestate')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='CollegeState', hue='Gender')
plt.xticks(rotation=90)  # Rotate x labels for better visibility
plt.title('Barplot of Gender by Collegestate')
plt.show()
```
